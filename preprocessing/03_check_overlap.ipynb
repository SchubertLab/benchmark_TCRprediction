# Load Data

```python
df_mutation = pd.read_csv(path_mutations, index_col=0)
df_mutation.head()
```

## Databases

```python
path_iedb = '../data/external/iedb_2808.csv'
path_vdjdb = '../data/external/vdjdb_2808.csv'
path_mcpas = '../data/external/mcpas_2808.csv'
```

```python
df_iedb = pd.read_csv(path_iedb, index_col=0)
df_iedb.head()
```

```python
df_vdjdb = pd.read_csv(path_vdjdb, index_col=0)
df_vdjdb = df_vdjdb.rename(columns={'antigen.epitope': 'epitope'})
df_vdjdb.head()
```

```python
df_mcpas = pd.read_csv(path_mcpas, index_col=0)
df_mcpas.head()
```

# Overlap to Mutation Dataset
Here, the combination of base receptor and epitope may be in the dataset, but not any mutations.

```python
base_epitopes = ['NLVPMVATV', 'VPSVWRSSL']
```

```python
tmp = df_iedb[df_iedb['IR_VDJ_1_junction_aa'].isin(df_mutation['CDR3_beta'])
             & (~df_iedb['epitope'].isin(base_epitopes))]
tmp
```

```python
tmp = df_vdjdb[df_vdjdb['IR_VDJ_1_junction_aa'].isin(df_mutation['CDR3_beta'])
              & (~df_vdjdb['epitope'].isin(base_epitopes))]
tmp
```

```python
tmp = df_mcpas[df_mcpas['IR_VDJ_1_junction_aa'].isin(df_mutation['CDR3_beta'])
               & (~df_mcpas['epitope'].isin(base_epitopes))]
tmp
```

We observe no overlap of the CDR3 beta chain to any relevant epitopes in the databases. Only the base epitopes or unrelated epitopes.
